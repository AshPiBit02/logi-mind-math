# Bisection Method

```python
import sympy as sp
import pandas as pd
df=pd.DataFrame(columns=["a","b","m","fm"])

# defining the symbol
x=sp.Symbol('x')

# defining the function
function=x**3-2*x-5
# print(function.subs(x,2)) usage


# Seed points
a=2
b=3
tolerance=0.0001


while abs(b-a)>tolerance:
    mid=(a+b)/2
    fm=float(function.subs(x,mid))
    df.loc[len(df)]=[a,b,mid,fm]
    if fm<0:
        a=mid
    else:
        b=mid
    
pd.options.display.float_format='{:.4f}'.format
   
print(df)

    

```
