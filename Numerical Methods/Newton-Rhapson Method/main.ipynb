# Newton-Rhapson Method

```python
import pandas as pd 
import sympy as sy 

df=pd.DataFrame(columns=["X_n","f(n)","f'(n)","X_n+1"])

x=sy.Symbol('x')

function=x**3-3*x-5
diff_f=sy.diff(function,x)
Xn=2
fx=float(function.subs(x,Xn)) # calculates f(n)
fdx=float(diff_f.subs(x,Xn)) # calculates f'(n)
Xn1=float(Xn-(fx/fdx)) # calculates X_(n+1)
tolerance=0.0001
while abs(Xn1-Xn)>tolerance:
    df.loc[len(df)]=[Xn,fx,fdx,Xn1]
    Xn=Xn1
    fx=float(function.subs(x,Xn)) # calculates f(n)
    fdx=float(diff_f.subs(x,Xn)) # calculates f'(n)
    Xn1=float(Xn-(fx/fdx)) # calculates X_(n+1)
pd.options.display.float_format='{:.4f}'.format
print(df)
```
