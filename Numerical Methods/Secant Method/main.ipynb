# Secant Method

```python
import pandas as pd
import sympy as sp
df=pd.DataFrame(columns=["X_n","X_n-1","f(Xn)","f(X_n-1)","X_n+1"])

x=sp.Symbol('x')
function=x**3-4*x-9 # Function

# Seed points
X_n=3 # Xn
X_n0=2 # Xn-1
f_x=float(function.subs(x,X_n)) # f(Xn)
f_x0=float(function.subs(x,X_n0)) # f(X_n-1)
X_n1=float(X_n-(f_x*((X_n-X_n0)/(f_x-f_x0)))) # X n+1
tolerance=0.0001

while abs(X_n-X_n0)>tolerance:
    df.loc[len(df)]=[X_n,X_n0,f_x,f_x0,X_n1]
    X_n0=X_n
    X_n=X_n1
    f_x=float(function.subs(x,X_n)) # f(Xn)
    f_x0=float(function.subs(x,X_n0)) # f(X_n-1)
    X_n1=float( X_n - (f_x * ( ( X_n - X_n0 ) / ( f_x - f_x0 ) ) ) )# X n+1

pd.options.display.float_format='{:.4f}'.format
print(df)

```
